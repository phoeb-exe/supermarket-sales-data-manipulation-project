```python
import pandas as pd
import numpy as np

#load dataset
df = pd.read_csv('supermarket_sales.csv')
df.head()
```

### 1. Filtering = Berapa banyak gross income yang di atas rata rata?

```python
#mencari mean
avg_gross = df['gross income'].mean()

#print filtered data
df[df['gross income'] > avg_gross]
```

```python
#print mean
print("Rata rata laba kotor (gross income) =", avg_gross)

#buat persentase
percentages = len(df[df['gross income'] > avg_gross]) /len(df) * 100

#print persentase
print(f"Banyaknya gross income di atas rata rata adalah {percentages}%")
```

Berdasarkan analisis di atas, ditemukan bahwa sekitar 40% transaksi memiliki gross income di atas rata-rata, sedangkan 60% sisanya berada di bawah rata-rata.

### 2. Sorting = Transaksi mana yang memiliki gross income paling tinggi dan paling rendah?

```python
#proses sorting
df.sort_values('gross income', ascending=False)
```

Dari tabel di atas terlihat bahwa transaksi yang memiliki gross income **paling tinggi** adalah transaksi dengan invoice ID 860-79-0874	dari branch C dengan product fashion accessories dengan income **49.6500**.
Sedangkan transaksi yang memiliki gross income **paling rendah** juga dari branch C dengan invoice ID 784-21-9238	dengan product sports and travel dengan income **0.5085**.

### 3. Grouping dan Aggregasi = Branch mana yang memiliki jumlah penjualan tertinggi?

```python
#proses membuat grouping dan aggregasi dari kolom branch dan quantity
grouped = df.groupby(['Branch']).agg({'Quantity': 'sum'})

#print data
grouped
```

Berdasarkan hasil grouping dan agregasi, branch A memiliki **jumlah penjualan tertinggi** jumlah penjualan **1859**. Hal ini menunjukkan bahwa branch A memiliki aktivitas penjualan yang paling tinggi dibanding cabang lainnya.

### 4. Join & Merging = Bagaimana hubungan antara gender dan product line setelah data transaksi digabungkan?

```python
#membuat df untuk customer data
df1 = df[['Invoice ID', 'Gender', 'Customer type']]

#membuat df untuk sales data
df2 = df[['Invoice ID', 'Product line', 'Quantity', 'Total']]

#menggabungkan kedua df berdasarkan invoice id dan menggunakan metode inner join
merge_df = pd.merge(df1, df2, on='Invoice ID', how='inner')

#grouping df berdasarkan product line dan gender
result = merge_df.groupby(['Product line', 'Gender']).size()
table = result.unstack(fill_value=0)

#print data
table
```

Dari hasil penggabungan di atas terlihat bahwa **gender female** lebih banyak membeli **product line fashion accessories**, sedangkan **gender male** lebih banyak membeli **product line electronic accessories**. Hal ini terjadi karena adanya preferensi produk yang berbeda berdasarkan gender.

### 5. Pivot = Bagaimana distribusi customer type di setiap branch?

```python
#membuat pivot table daru index Branch dengan kolom Customer type dan value Invoice ID yang dijumlahkan/count
pivot_table = pd.pivot_table(df, index='Branch', columns='Customer type', values='Invoice ID', aggfunc='count', fill_value=0)

#print data
pivot_table
```

Berdasarkan hasil pivot, **branch A** memiliki **pelanggan normal paling banyak (173)**, sedangkan **branch B** memiliki **jumlah pelanggan member dan normal yang hampir sama**. Hal ini menunjukkan bahwa loyalitas member di **branch C paling tinggi** dibanding branch lain.

### 6. Crosstab Analysis = Bagaimana distribusi customer type terhadap product line?

```python
#membuat crosstab dari index Product line dan kolom Customer type
crosstab_analysis = pd.crosstab(index=df['Product line'], columns=df['Customer type'])

#print data
crosstab_analysis
```

Dari hasil crosstab, **pelanggan normal** lebih banyak membeli **Electronic accessories	dan Fashion accessories**	dengan jumlah **92 transaksi**, sedangkan **pelanggan member** lebih banyak membeli **Food and beverages**.

### 8. Datetime Operations = Bagaimana tren quantity per bulan selama periode?

```python
#mengubah kolom date menjadi datetime
df['Date'] = pd.to_datetime(df['Date'])

#membuat dataframe baru untuk month dengan mengambil nilai month pada kolom data
df['Month'] = df['Date'].dt.to_period('M')
```

```python
#membuat monthly tren dari groupping
monthly_tren = df.groupby('Month')['Quantity'].sum().reset_index()
print(monthly_tren)

#mengubah ke bentuk string agar dapat divisualisasikan
monthly_tren['Month'] = monthly_tren['Month'].astype(str)
```

```python
import matplotlib.pyplot as plt

#membuat visualisasi tren
plt.plot(monthly_tren['Month'], monthly_tren['Quantity'])
plt.title('Tren Quantity per Bulan')
plt.xlabel('Bulan')
plt.ylabel('Total Quantity')

plt.show()
```

Dari visual di atas, bulan Januari adalah bulan dengan jumlah penjualan (secara kuantitas) paling banyak tetapi pada bulan Februari jumlah penjualan menurun drastis tetapi stabil kembali di bulan Maret.

# 2. Standar Scaller for Exploratory Data Analysis ?
`clue` : lakukan tanpa splitting dataset ke train dan test
- lakukan pada kolom Total, Unit Price, dan tambahkan 2 kolom lainnya
- dan tulis insight dari yang anda dapat dari grafik tersebut

```python
#import seaborn dan scaler
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler

#memanggil scaler
standard_scaler = StandardScaler()

#membuat kolom untuk standard scaler
standard_cols = ['Total', 'Unit price', 'Quantity', 'Rating']

#proses standard scaler
df[standard_cols] = standard_scaler.fit_transform(df[standard_cols])
```

```python
#cek distribusi datanya
num_plots = len(standard_cols)
ncols = 2
nrows = (num_plots + ncols - 1) // ncols

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5 * nrows))

for i, feature in enumerate(standard_cols):
  row = i // ncols
  col = i % ncols
  sns.histplot(df[feature], ax=axes[row, col])
  axes[row, col].set_title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

Dari distribusi di atas, grafik Total tetap miring ke kanan (right-skewed) meskipun sudah distandardisasi dan untuk grafik Unit price, Quantity, dan Rating terlihat relatif lebih merata, yang menunjukkan bahwa persebaran nilai yang tidak ekstrem.

# 3. MinMax Scaller for Exploratory Data Analysis ?
`clue` : lakukan tanpa splitting dataset ke train dan test
- lakukan pada kolom Total, Unit Price, dan tambahkan 2 kolom lainnya
- dan tulis insight dari yang anda dapat dari grafik tersebut

```python
#memanggil scaler
minmax_scaler = MinMaxScaler()

#membuat kolom untuk minmax scaler
minmax_cols = ['Total', 'Unit price', 'cogs', 'gross income']

#proses minmax scaler
df[minmax_cols] = minmax_scaler.fit_transform(df[minmax_cols])
```

```python
#cek distribusi datanya
num_plots = len(minmax_cols)
ncols = 2
nrows = (num_plots + ncols - 1) // ncols

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, 5 * nrows))

for i, feature in enumerate(minmax_cols):
  row = i // ncols
  col = i % ncols
  sns.histplot(df[feature], ax=axes[row, col])
  axes[row, col].set_title(f'Distribution of {feature}')

plt.tight_layout()
plt.show()
```

Dari distribusi di atas, gratik Total, cogs, dan gross income masih memperlihatkan bentuk distribusi yang miring ke kanan yang menunjukkan bahwa sebagian besar transaksi bernilai kecil. Sedangkan grafuik Unit price relatif lebih seragam.
